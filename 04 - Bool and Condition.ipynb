## Conditions

Python语言提供了if-else和if-elif-else语句作为条件判断：



```python
number = 10
guess_number = int(input("Please enter a number here: "))

if guess_number > 10:
    print("Your number is bigger.")
elif guess_number < 10:
    print("Your number is smaller.")
else:
    print("Your number is correct.")
```

需要注意：else后面不需要任何的判断条件，因为else表示的是除了上面之外的所有的情况。
