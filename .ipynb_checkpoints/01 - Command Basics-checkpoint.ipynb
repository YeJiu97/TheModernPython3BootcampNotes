## 路径

计算机都是按照目录或者文件来进行组织，然后是文件。

最顶级的一个目录，被称之为叫做 root directory（root /）。这个目录在计算机的某一个位置，然后在这个目录之下，还有着树状结构的其他的文件夹：



如果想要知道自己现在在哪个位置，可以运行pwd指令：
```
pwd
```
然后会返回当前所在的文件夹的路径：

```
yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/Github/TheModernPython3BootcampNotes
$ pwd
/c/Users/yinia/Desktop/Github/TheModernPython3BootcampNotes
```

我们可以使用cd来进入或者退出不同的文件夹：
```
yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ pwd
/c/Users/yinia/Desktop/123

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ cd 1

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123/1
$ pwd
/c/Users/yinia/Desktop/123/1

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123/1
$ cd ..
```
在上述的代码中，还使用了ls，ls可以将当前文件夹之下的文件和文件夹中的内容都罗列出来，如果存在着一些隐藏的文件，则需要给定一个参数 -a 才能够打印出来：
```
ls -a
```

或者我们可以在不改变当前的文件夹的情况之下打印某一个文件夹中的内容：
```
$ pwd
/c/Users/yinia/Desktop/123/1

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123/1
$ ls
1.txt

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123/1
$ cd ..

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls ./1
1.txt

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ pwd
/c/Users/yinia/Desktop/123
```

## 创建，移动和删除

如果要创建文件夹，则可以使用mkdir指令来实现：
```
yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ mkdir 3

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/  3/
```
如果要创建的不是文件夹，而是具体的文件，则需要使用touch来实现：
```
yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/  3/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ touch 2.py

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/  2.py  3/
```
如果要移动文件，则可以使用mv指令来进行实现：
```
yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/  2.py  3/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ mv 2.py ./2/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/  3/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls ./2
2.py
```
mv指令也可以对文件进行重命名：
```
yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/  3/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls ./2
2.py

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ mv ./2/2.py ./3/3.py

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/  3/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls ./2

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls ./3
3.py
```

如果要删除文件，可以使用rm来进行实现：
```
yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/  3/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ cd 3

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123/3
$ ls
3.py

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123/3
$ rm 3.py

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123/3
$ ls
```

rm并不能够用来直接删除文件夹：
```
yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/  3/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ rm 3
rm: cannot remove '3': Is a directory
```

如果要删除directory，则需要提供 -r 来作为参数：
```
yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/  3/

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ rm 3
rm: cannot remove '3': Is a directory

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ rm -r 3

yinia@DESKTOP-S9ODFEQ MINGW64 ~/Desktop/123
$ ls
1/  2/
```
