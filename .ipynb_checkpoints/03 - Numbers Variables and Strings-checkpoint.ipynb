## Numbers and Operators

Python语言提供了三种Number的类型：
1. int
2. float
3. complex
其中int和float是常用的两种类型。

```python
a = 30  # 这是一个int类型
b = 30.15  # 这是一个float类型
```

```python
a
```

```python
b
```

```python
type(a)  # type()函数可以将括号中的变量的类型打印出来
```

```python
type(b)
```

```python
type(1 + 1.3)
```

对于Number的基础的操作如下所示：



```python
a = 10
b = 20
```

```python
a + b
```

```python
a - b
```

```python
a * b
```

```python
a / b
```

```python
a ** b
```

```python
a % b  # 求余数
```

```python
a // b  # 求整除的商
```

## Variables and Strings

变量是存储在内存中的值，这就意味着在创建变量时会在内存中开辟一个空间。

Python语言不像Java或者C++之类的语言，不需要在赋值的时候对类型进行声明。

Python语言在赋值的时候采用 = 来进行，如果是判断等式的两边是否相等，则使用 == 来进行判断。

Python语言的命名规则如下所示：
1. 只能够以字母或者_来开头
2. 名字不能够包含空格
3. 名字不能够是Python语言的关键字

```python
# 以下是一些例子
a = 10
_a = 20
string_name = 30
StringName = 40
string_name_5 = 50
StringName6 = 60
```
