## Python 安装

首先可以使用：
```
python
```
来确认自己的电脑是否已经安装了Python语言：
```
C:\Users\yinia>python
Python 3.10.5 (tags/v3.10.5:f377153, Jun  6 2022, 16:14:13) [MSC v.1929 64 bit (AMD64)] on win32
Type "help", "copyright", "credits" or "license" for more information.
```

可以发现电脑上已经安装了Python，并且版本为3.10.5。

如果没有安装Python，则需要找到Python语言的官网来进行安装：https://www.python.org/ 。



不推荐直接下载最新版本的Python，因为存在着一些第三方库不支持的情况。

## 编辑器

Python语言会有自带的IDLE：



通过使用Python自带的IDLE可以进行更加方便的写代码。

但是Python自带的IDLE不如一些第三方的编辑器来的好用，这里推荐的是：
1. Sublime
2. Visual Studio Code

此外，推荐的编译器是Pycharm。

编辑器和编译器的区别在于：
> 编辑器：即编写代码的工具，方便人对程序的操作和管理如窗口系统中的记事本以及Linux的系统中的VI / vim的都是编辑器。 编译器：将高级计算机语言程序翻译为计算机能解读，运行的低级机器语言程序。
